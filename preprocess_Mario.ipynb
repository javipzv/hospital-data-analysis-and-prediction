## ¿Columnas redundantes?

En primer lugar, vemos que hay 2 columnas asociadas al ID del paciente. Veamos si la información que contienen es exactamente igual o difieren en alguna fila.

```python
data1.loc[data1["patient ID"] != data1["patient ID.1"]]
```

Hay 7 filas donde ambas columnas difieren. Comprobemos una cosa:

```python
ids1 = data1.loc[data1["patient ID"] != data1["patient ID.1"]]['patient ID.1']

data1.loc[data1["patient ID"].isin(ids1)]
```

En segundo lugar, vemos que existe una columna con un nombre extravagante, cuyos valores parecen iguales a los de la columna que define la fecha de los primeros síntomas "date_of_first_symptoms"

```python
data1.loc[(data1['date_of_first_symptoms'] != data1['BASVURUTARIHI'])]
```

Como vemos, no hay ninguna columna donde haya valores diferentes, por tanto, podemos decir con certeza que la columna BASVURUTARIHI es igual a la anterior. Por tanto, procedemos a eliminarla.

```python
data1.drop(['BASVURUTARIHI'], axis=1, inplace=True)
data1.drop(['patient ID.1'], axis=1, inplace=True)
```

#### Tratamiento de nulos

```python
data_null = data1.isna().sum()[data1.isna().sum() > 0].index
data_null
```

En primer lugar, eliminamos las filas donde PCR_Result sea NaN, ya que es la variable a predecir.

```python
data1.dropna(subset=['PCR_result'], inplace=True)
```

Tras esto, vemos que la siguiente columna con un mayor volumen de nulos es la fiebre. En este caso, vamos a hacer una regresión para imputar valores, en los datos faltantes.

```python
# Remove non-numeric columns
numeric_data1 = data1.select_dtypes(include=[np.number])

# Calculate correlation and get the top 5 variables related to 'fever_temperature'
numeric_data1.corr()['fever_temperature'].abs().sort_values(ascending=False).head(6)
```

Viendo estos resultados de correlación, pensamos que no tiene mucho sentido realizar una regresión para la imputación de valores, en cambio, hemos decidido que la mejor opción de imputación será mediante la media.

```python
# do a regression imputation on fever_temperature
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Split the data into train and test
```

```python
#import simple imputer
from sklearn.impute import SimpleImputer

# Create an imputer object with a median filling strategy
imputer = SimpleImputer(strategy='mean')

data1['fever_temperature'] = imputer.fit_transform(data1[['fever_temperature']])
```

Adicionalmente a esto, vemos que hay algunas columnas binarias de síntomas que tienen muy pocos valores nulos. En nuestro caso, vamos a sustituir esos valores nulos por 0, ya que son columnas binarias donde 1 indica que tiene ese síntoma y 0 que no lo ha tenido. Pensamos que si el valor es nulo en esas columnas, significa que no se han hecho las pruebas para determinar el síntoma y, por tanto, suponemos que no ha tenido ese síntoma.


```python
data1[data_null[2:]] = data1[data_null[2:]].fillna(0)
```

```python
data1.isna().sum()
```

## IMPORTANTE: Añadir los -1 de oxygen_saturation a nulos para imputarlos.

Ahora, solo nos quedan valores nulos en oxygen saturation, en este caso es un caso parecido al de la fiebre, haremos una imputación basada en la media

```python
data1['oxygen_saturation'] = imputer.fit_transform(data1[['oxygen_saturation']])
```

Tras estas transformaciones, hemos eliminado todos los valores nulos de este dataset. 

### Variables Categóricas

En esta nueva sección, lo que vamos a realizar es tratar con las variables categóricas que tenemos en nuestro conjunto de datos. En primer lugar, vamos a tratar la columna de nacionalidad, donde vemos que hay muchos valores con nombre T.C, en un inicio no sabíamos que significaba, pero teniendo en cuenta que teníamos un nombre de columna en turco, entonces desciframos que esas siglas significan Turquía. Por tanto, vamos a sustituirlo.

```python
#Cambia T.C en Turquia en la columna nationality

data1['nationality'] = data1['nationality'].replace('T.C.', 'Turkish')

```

```python
data1['nationality'].value_counts()
```

Tras haber hecho esto, las otras 2 columnas categóricas que son el género y el resultado de la prueba de la PCR, vamos a codificarlas en variables numéricas.

```python
from sklearn.preprocessing import LabelEncoder

# Create a label encoder object
le = LabelEncoder()

# Fit the encoder to the pandas column
data1['gender K=female E=male'] = le.fit_transform(data1['gender K=female E=male'])
```

```python
data1['PCR_result'] = le.fit_transform(data1['PCR_result'])
```

```python
data1.rename(columns={'gender K=female E=male': 'gender'}, inplace=True)

          
```
